```python
MEL_FIXED = 128  # zaman ekseni sabit boyut

def extract_features(file_path, n_mfcc=40, n_mels=128, duration=30):
    try:
        y, sr = librosa.load(file_path, duration=duration, mono=True)

        # MFCC
        mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
        mfcc_mean = np.mean(mfcc, axis=1)
        mfcc_std = np.std(mfcc, axis=1)

        # Mel-Spectrogram — sabit boyuta kırp veya pad et
        mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=n_mels)
        mel_db = librosa.power_to_db(mel, ref=np.max)

        # Zaman eksenini 128'e sabitle
        if mel_db.shape[1] >= MEL_FIXED:
            mel_db = mel_db[:, :MEL_FIXED]
        else:
            pad = MEL_FIXED - mel_db.shape[1]
            mel_db = np.pad(mel_db, ((0,0),(0,pad)), mode='constant')

        # Ek özellikler
        chroma = np.mean(librosa.feature.chroma_stft(y=y, sr=sr), axis=1)
        zcr = np.mean(librosa.feature.zero_crossing_rate(y))
        spectral_centroid = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr))

        return {
            'mfcc_mean': mfcc_mean,
            'mfcc_std': mfcc_std,
            'mel': mel_db,
            'chroma': chroma,
            'zcr': zcr,
            'spectral_centroid': spectral_centroid
        }
    except Exception as e:
        print(f"Hata: {file_path} — {e}")
        return None
```

```python
GTZAN_PATH = '/content/drive/MyDrive/music_genre_project/data/Data/genres_original/'
TR_RAP_PATH = '/content/drive/MyDrive/music_genre_project/data/turkish_rap/'
SAVE_PATH = '/content/drive/MyDrive/music_genre_project/data/'

genres_gtzan = ['blues', 'classical', 'country', 'disco',
                'hiphop', 'jazz', 'metal', 'pop', 'reggae', 'rock']

all_mfcc = []
all_mel = []
all_labels = []
errors = []

# GTZAN verisi
for label, genre in enumerate(genres_gtzan):
    folder = os.path.join(GTZAN_PATH, genre)
    files = glob.glob(folder + '/*.wav')
    print(f"[{label+1}/11] {genre}: {len(files)} dosya işleniyor...")
    for f in files:
        feat = extract_features(f)
        if feat:
            mfcc_vec = np.concatenate([feat['mfcc_mean'], feat['mfcc_std'],
                                        feat['chroma'], [feat['zcr'], feat['spectral_centroid']]])
            all_mfcc.append(mfcc_vec)
            all_mel.append(feat['mel'])
            all_labels.append(label)
        else:
            errors.append(f)

# Türkçe rap verisi — label 10
print(f"[11/11] turkish_rap: {len(glob.glob(TR_RAP_PATH + '*.wav'))} dosya işleniyor...")
for f in glob.glob(TR_RAP_PATH + '*.wav'):
    feat = extract_features(f)
    if feat:
        mfcc_vec = np.concatenate([feat['mfcc_mean'], feat['mfcc_std'],
                                    feat['chroma'], [feat['zcr'], feat['spectral_centroid']]])
        all_mfcc.append(mfcc_vec)
        all_mel.append(feat['mel'])
        all_labels.append(10)
    else:
        errors.append(f)

# Kaydet
np.save(SAVE_PATH + 'X_mfcc.npy', np.array(all_mfcc))
np.save(SAVE_PATH + 'X_mel.npy', np.array(all_mel))
np.save(SAVE_PATH + 'y_labels.npy', np.array(all_labels))

print(f"\nTamamlandı!")
print(f"MFCC shape: {np.array(all_mfcc).shape}")
print(f"Mel shape: {np.array(all_mel).shape}")
print(f"Toplam örnek: {len(all_labels)}")
print(f"Hatalı dosya: {len(errors)}")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.preprocessing import LabelEncoder

# Mel veriyi yükle ve CNN için hazırla
X_mel = np.load(SAVE_PATH + 'X_mel.npy')
y = np.load(SAVE_PATH + 'y_labels.npy')

# CNN için boyut ekle (128, 128, 1)
X_mel = X_mel[..., np.newaxis]

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(
    X_mel, y, test_size=0.2, random_state=42, stratify=y
)

print(f"Train shape: {X_train_m.shape}")
print(f"Test shape: {X_test_m.shape}")

# CNN modeli
def build_cnn(input_shape, n_classes=11):
    model = models.Sequential([
        layers.Input(shape=input_shape),

        layers.Conv2D(32, (3,3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2,2)),

        layers.Conv2D(64, (3,3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2,2)),

        layers.Conv2D(128, (3,3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.GlobalAveragePooling2D(),

        layers.Dense(256, activation='relu'),
        layers.Dropout(0.4),
        layers.Dense(n_classes, activation='softmax')
    ])
    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

model = build_cnn(X_train_m.shape[1:])
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

MODEL_PATH = '/content/drive/MyDrive/music_genre_project/models/'
import os
os.makedirs(MODEL_PATH, exist_ok=True)

callbacks = [
    EarlyStopping(patience=5, restore_best_weights=True, verbose=1),
    ModelCheckpoint(MODEL_PATH + 'best_model.keras', save_best_only=True, verbose=0)
]

history = model.fit(
    X_train_m, y_train_m,
    validation_split=0.2,
    epochs=50,
    batch_size=32,
    callbacks=callbacks,
    verbose=1
)

# Test sonucu
test_loss, test_acc = model.evaluate(X_test_m, y_test_m, verbose=0)
print(f"\nCNN Test Accuracy: {test_acc:.4f}")
```

```python
from tensorflow.keras import layers, models
import tensorflow as tf

def build_cnn_v2(input_shape, n_classes=11):
    model = models.Sequential([
        layers.Input(shape=input_shape),

        # Augmentation (sadece eğitimde aktif)
        layers.RandomFlip("horizontal"),
        layers.RandomZoom(0.1),

        layers.Conv2D(32, (3,3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2,2)),
        layers.Dropout(0.2),

        layers.Conv2D(64, (3,3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2,2)),
        layers.Dropout(0.2),

        layers.Conv2D(128, (3,3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2,2)),
        layers.Dropout(0.2),

        layers.GlobalAveragePooling2D(),

        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(n_classes, activation='softmax')
    ])
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

model_v2 = build_cnn_v2(X_train_m.shape[1:])

history_v2 = model_v2.fit(
    X_train_m, y_train_m,
    validation_split=0.2,
    epochs=50,
    batch_size=16,  # batch size küçültüldü
    callbacks=callbacks,
    verbose=1
)

test_loss, test_acc = model_v2.evaluate(X_test_m, y_test_m, verbose=0)
print(f"\nCNN v2 Test Accuracy: {test_acc:.4f}")
```

```python
from sklearn.ensemble import GradientBoostingClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, accuracy_score

X_mfcc = np.load(SAVE_PATH + 'X_mfcc.npy')
y = np.load(SAVE_PATH + 'y_labels.npy')

X_train, X_test, y_train, y_test = train_test_split(
    X_mfcc, y, test_size=0.2, random_state=42, stratify=y
)

genres_all = ['blues','classical','country','disco','hiphop',
              'jazz','metal','pop','reggae','rock','turkish_rap']

# Deney 1 — SVM (küçük veri setinde genellikle en iyi)
svm = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC(kernel='rbf', C=10, gamma='scale', probability=True))
])
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
acc_svm = accuracy_score(y_test, y_pred_svm)
print(f"SVM Accuracy: {acc_svm:.4f}")
print(classification_report(y_test, y_pred_svm, target_names=genres_all))
```

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

results = {}
results['Random Forest (baseline)'] = 0.6909
results['CNN v1'] = 0.5182
results['CNN v2'] = 0.4860
results['SVM (RBF, C=10)'] = acc_svm

# Deney 4 — SVM farklı parametreler
svm_v2 = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC(kernel='rbf', C=100, gamma='scale'))
])
svm_v2.fit(X_train, y_train)
acc_svm_v2 = accuracy_score(y_test, svm_v2.predict(X_test))
results['SVM (RBF, C=100)'] = acc_svm_v2
print(f"SVM C=100 Accuracy: {acc_svm_v2:.4f}")

# Deney 5 — Voting Classifier (RF + SVM birlikte)
voting = VotingClassifier(estimators=[
    ('rf', RandomForestClassifier(n_estimators=200, random_state=42)),
    ('svm', Pipeline([('scaler', StandardScaler()), ('svm', SVC(kernel='rbf', C=10, gamma='scale', probability=True))]))
], voting='soft')
voting.fit(X_train, y_train)
acc_voting = accuracy_score(y_test, voting.predict(X_test))
results['Voting (RF + SVM)'] = acc_voting
print(f"Voting Classifier Accuracy: {acc_voting:.4f}")

# Sonuçları yazdır
print("\n===== EK 1: BAŞARIM KARŞILAŞTIRMA TABLOSU =====")
for model_name, acc in results.items():
    bar = "█" * int(acc * 20)
    print(f"{model_name:<30} {acc:.4f}  {bar}")
```

```python
import os
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

SAVE_PATH = "/content/drive/MyDrive/music_genre_project/data/"

X_mfcc = np.load(SAVE_PATH + "X_mfcc.npy")
y_labels = np.load(SAVE_PATH + "y_labels.npy")

genres_all = [
    "blues",
    "classical",
    "country",
    "disco",
    "hiphop",
    "jazz",
    "metal",
    "pop",
    "reggae",
    "rock",
    "turkish_rap"
]

X_train, X_test, y_train, y_test = train_test_split(
    X_mfcc,
    y_labels,
    test_size=0.2,
    random_state=42,
    stratify=y_labels
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
svm_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(kernel="rbf"))
])

param_grid = {
    "svm__C": [1, 10, 100],
    "svm__gamma": ["scale", 0.01, 0.001]
}

grid_search = GridSearchCV(
    svm_pipeline,
    param_grid,
    cv=5,
    scoring="accuracy",
    n_jobs=-1,
    verbose=2
)

grid_search.fit(X_train, y_train)

best_svm = grid_search.best_estimator_
y_pred_svm_grid = best_svm.predict(X_test)

grid_acc = accuracy_score(y_test, y_pred_svm_grid)

print("Best parameters:", grid_search.best_params_)
print("Best CV accuracy:", grid_search.best_score_)
print("Test accuracy:", grid_acc)

print("\nClassification Report:")
print(classification_report(y_test, y_pred_svm_grid, target_names=genres_all))
```

```python
from sklearn.ensemble import RandomForestClassifier, VotingClassifier

feature_sets = {
    "İlk 40 özellik": X_mfcc[:, :40],
    "İlk 60 özellik": X_mfcc[:, :60],
    "Tüm 94 özellik": X_mfcc
}

ablation_results = []

for name, X_feat in feature_sets.items():
    X_train_f, X_test_f, y_train_f, y_test_f = train_test_split(
        X_feat,
        y_labels,
        test_size=0.2,
        random_state=42,
        stratify=y_labels
    )

    svm_model = Pipeline([
        ("scaler", StandardScaler()),
        ("svm", SVC(kernel="rbf", C=10, gamma="scale", probability=True))
    ])

    rf_model = RandomForestClassifier(
        n_estimators=200,
        random_state=42
    )

    voting_model = VotingClassifier(
        estimators=[
            ("rf", rf_model),
            ("svm", svm_model)
        ],
        voting="soft"
    )

    voting_model.fit(X_train_f, y_train_f)
    y_pred_f = voting_model.predict(X_test_f)
    acc = accuracy_score(y_test_f, y_pred_f)

    ablation_results.append({
        "Öznitelik Seti": name,
        "Öznitelik Sayısı": X_feat.shape[1],
        "Accuracy": acc
    })

ablation_df = pd.DataFrame(ablation_results)
print(ablation_df)

ablation_df.to_csv(SAVE_PATH + "ablation_results.csv", index=False)
print("Ablation sonuçları kaydedildi:", SAVE_PATH + "ablation_results.csv")
```
