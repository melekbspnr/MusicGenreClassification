```python
!pip -q install google-genai

from google import genai
import os

# Drive kullanıyorsan mount et
from google.colab import drive
drive.mount('/content/drive')

print("GEMINI_API_KEY var mı?", "GEMINI_API_KEY" in os.environ)

client = genai.Client()
MODEL_NAME = "gemini-2.5-pro"

def call_agent(agent_name, role, task, context):
    prompt = f"""
Sen çok etmenli bir Agentic AI sistemindeki '{agent_name}' adlı ajansın.

Rolün:
{role}

Görevin:
{task}

Bağlam / Kullanıcı girdisi:
{context}

Kurallar:
- Türkçe cevap ver.
- Kısa, net ve profesyonel yaz.
- Kullanıcının vermediği deneyimleri uydurma.
- Abartılı, gerçek dışı veya kanıtlanmamış ifadeler kullanma.
- Çıktını başlıklar ve maddeler halinde düzenle.
"""
    response = client.models.generate_content(
        model=MODEL_NAME,
        contents=prompt
    )
    return response.text


user_input = """
Hedef pozisyon: Yazılım geliştirme / yapay zeka / veri bilimi stajı

Kullanıcının becerileri:
Python, makine öğrenmesi, temel derin öğrenme, Java, SQL, Git, Google Colab

Projeler:
1) Music Genre Classification projesi:
GTZAN veri seti ve özel Turkish Rap sınıfı kullanıldı.
MFCC, Mel-Spectrogram, Chroma, Zero Crossing Rate ve Spectral Centroid öznitelikleri çıkarıldı.
Random Forest, SVM, CNN ve Voting Classifier modelleri karşılaştırıldı.
En iyi sonuç Voting Classifier ile %75 accuracy oldu.
Turkish Rap sınıfında yüksek recall elde edildi.

2) Ranorex masaüstü test otomasyonu staj deneyimi:
Masaüstü uygulamalarında test otomasyonu, repository kullanımı, data-driven testing,
validation ve raporlama konularında çalışma yapıldı.

İstenen çıktılar:
- Profil analizi
- Hedef pozisyona uygunluk analizi
- CV için proje açıklaması
- Kısa ve istekli staj başvuru maili
- Eksik yetenek ve geliştirme önerileri
"""


profile_output = call_agent(
    "Profil Analiz Ajanı",
    "Kullanıcının beceri, proje ve deneyimlerini analiz eden kariyer ajanı.",
    "Kullanıcının güçlü teknik yönlerini ve öne çıkarılabilecek deneyimlerini belirle.",
    user_input
)

fit_output = call_agent(
    "Pozisyon Uyum Ajanı",
    "Kullanıcı profilini hedef staj pozisyonu ile eşleştiren ajan.",
    "Yazılım geliştirme, yapay zeka ve veri bilimi stajı açısından kullanıcının uygunluğunu değerlendir.",
    user_input + "\n\nProfil Analizi:\n" + profile_output
)

cv_output = call_agent(
    "CV Geliştirme Ajanı",
    "Projeleri ve deneyimleri CV'de kullanılabilecek profesyonel açıklamalara dönüştüren ajan.",
    "Kullanıcının projelerini CV'ye yazılacak 4-6 güçlü madde haline getir.",
    user_input + "\n\nProfil Analizi:\n" + profile_output + "\n\nPozisyon Uyum Analizi:\n" + fit_output
)

mail_output = call_agent(
    "Başvuru Maili Ajanı",
    "İK veya teknik ekip için kısa, profesyonel ve istekli başvuru maili oluşturan ajan.",
    "Kullanıcının staj başvurusu için kısa ve istekli bir mail taslağı oluştur.",
    user_input + "\n\nProfil Analizi:\n" + profile_output + "\n\nCV Açıklamaları:\n" + cv_output
)

critic_output = call_agent(
    "Eleştirmen ve Doğruluk Kontrol Ajanı",
    "Üretilen metinleri kontrol eden, abartılı veya kanıtsız ifadeleri düzelten ajan.",
    "Tüm çıktıları gerçekçilik, profesyonellik ve doğruluk açısından değerlendir. Gerekli düzeltmeleri öner.",
    user_input
    + "\n\nProfil Analizi:\n" + profile_output
    + "\n\nPozisyon Uyum Analizi:\n" + fit_output
    + "\n\nCV Çıktısı:\n" + cv_output
    + "\n\nMail Çıktısı:\n" + mail_output
)

final_output = call_agent(
    "Final Rapor Ajanı",
    "Tüm ajan çıktılarını birleştirip son kullanıcıya düzenli final raporu sunan ajan.",
    "Profil analizi, uygunluk analizi, CV maddeleri, başvuru maili ve geliştirme önerilerini nihai rapor olarak düzenle. Eleştirmen ajan önerilerini dikkate al.",
    user_input
    + "\n\nProfil Analizi:\n" + profile_output
    + "\n\nPozisyon Uyum Analizi:\n" + fit_output
    + "\n\nCV Çıktısı:\n" + cv_output
    + "\n\nMail Çıktısı:\n" + mail_output
    + "\n\nEleştirmen Çıktısı:\n" + critic_output
)

print("===== PROFİL ANALİZ AJANI =====")
print(profile_output)

print("\n===== POZİSYON UYUM AJANI =====")
print(fit_output)

print("\n===== CV GELİŞTİRME AJANI =====")
print(cv_output)

print("\n===== BAŞVURU MAİLİ AJANI =====")
print(mail_output)

print("\n===== ELEŞTİRMEN AJAN =====")
print(critic_output)

print("\n===== FİNAL RAPOR =====")
print(final_output)


SAVE_PATH = "/content/drive/MyDrive/music_genre_project/data/"
os.makedirs(SAVE_PATH, exist_ok=True)

output_file = os.path.join(SAVE_PATH, "ek5_agentic_ai_staj_asistani.txt")

with open(output_file, "w", encoding="utf-8") as f:
    f.write("ÇOK ETMENLİ LLM DESTEKLİ STAJ BAŞVURU ASİSTANI\n\n")
    f.write("KULLANICI GİRDİSİ:\n" + user_input + "\n\n")
    f.write("===== PROFİL ANALİZ AJANI =====\n" + profile_output + "\n\n")
    f.write("===== POZİSYON UYUM AJANI =====\n" + fit_output + "\n\n")
    f.write("===== CV GELİŞTİRME AJANI =====\n" + cv_output + "\n\n")
    f.write("===== BAŞVURU MAİLİ AJANI =====\n" + mail_output + "\n\n")
    f.write("===== ELEŞTİRMEN AJAN =====\n" + critic_output + "\n\n")
    f.write("===== FİNAL RAPOR =====\n" + final_output + "\n")

print("\nDosya kaydedildi:", output_file)
```
